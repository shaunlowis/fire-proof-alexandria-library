## Places that I have lived so far
- Written in python 3.8.6
- Modules used; matplotlib and cartopy

```python
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature
from matplotlib.font_manager import FontProperties
```

```python
def add_text(x, y, text, **kwargs):
    """ Helper function to avoid having to repeat the same arguments when adding locations to the map. """
    ax.text(x, y, text, transform=ccrs.PlateCarree(), verticalalignment='top', fontsize=14, c='red', bbox={'facecolor': 'black', 'alpha': 0.8, 'pad': 3}, **kwargs)
```

```python
proj = cartopy.crs.PlateCarree()
fig, ax = plt.subplots(1, figsize=(20,20), sharey=True, subplot_kw=dict(projection=proj))

locations = {'Vorchdorf, Austria': (47.999895203735676, 13.922198455778863), 'Stilbay, South-Africa': (-34.37366352215708, 21.41477988683966), 
             'Alexandra, New Zealand': (-45.24890826338315, 169.3897437530176), 'Christchurch, New Zealand': (-43.53138878592723, 172.58813845375494),
             'Paraparaumu, New Zealand': (-40.92248793152586, 174.9790291769173)}

keys = [key for key in locations]

for i, key in enumerate(keys):
    ax.plot(locations[key][1], locations[key][0], transform=ccrs.PlateCarree(), color='red', linewidth=4, marker='x')

# I have to adjust the locations of the text labels manually here
# Austria:
add_text(locations[keys[0]][1] + 2, locations[keys[0]][0] + 2, keys[0])
# South-Africa
add_text(locations[keys[1]][1] - 1, locations[keys[1]][0] - 2, keys[1])
# Alexandra
add_text(locations[keys[2]][1] - 2, locations[keys[2]][0] + 1, keys[2], horizontalalignment='right')
# Christchurch
add_text(locations[keys[3]][1] - 2, locations[keys[3]][0] + 3, keys[3], horizontalalignment='right')
# Wellington
add_text(locations[keys[4]][1] - 2, locations[keys[4]][0] + 4, keys[4], horizontalalignment='right')

ax.coastlines()
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.RIVERS)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')
    
plt.title('Places I have lived so far', fontsize=20)

plt.show()
```
